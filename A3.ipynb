```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
```

```python
my_pivot = pd.pivot_table(df, 
                          index = 'desc_ent' ,
                          values = 'peso'
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df, 
                          index = 'desc_ent' ,
                          columns = 'sexo' ,
                          values = 'peso'
                         )
my_pivot
```

En estas tablas podemos observar el peso promedio de cada estado así como el peso promedio de cada genero por estado. Esto se consiguió realizando una tabla de pivote donde se escogieron las unidades deseadas.
